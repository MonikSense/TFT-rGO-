```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import VarianceThreshold
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score

# Load dataset
url = "https://raw.githubusercontent.com/MonikSense/TFT-rGO-/refs/heads/main/TFT%20design%20ML%20FG3.CSV"
df = pd.read_csv(url)

# Drop rows with any NA in target (optional: if target column has NAs)
df.dropna(subset=["VT(V)"], inplace=True)  # Replace "ID" with your target column name

# Handle missing values
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
imputer = SimpleImputer(strategy='mean')
df[num_cols] = imputer.fit_transform(df[num_cols])

# Feature selection: variance threshold
selector = VarianceThreshold(threshold=0.01)
df_var = selector.fit_transform(df[num_cols])
selected_columns = num_cols[selector.get_support()]
df = df[selected_columns]



# Set target and features
target_col = 'VT(V)'  # Replace with actual target column
X = df.drop(columns=[target_col], errors='ignore')
y = df[target_col]

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42)

# Define model and hyperparameters
rfr = RandomForestRegressor(random_state=42)
param_grid = {
    'n_estimators': [200, 500],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# Grid Search with 5-fold cross-validation
grid_search = GridSearchCV(estimator=rfr, param_grid=param_grid,
                           cv=5, n_jobs=-1, scoring='r2')
grid_search.fit(X_train, y_train)

# Best model and results
best_model = grid_search.best_estimator_
print("Best parameters:", grid_search.best_params_)

# Predictions and evaluation
y_pred = best_model.predict(X_test)
print("RÂ² score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))

# Plot Predicted vs Actual
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='navy', alpha=0.6, edgecolors='k')
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         color='red', lw=2, linestyle='--')
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Random Forest: Actual vs Predicted")
plt.grid(True)
plt.tight_layout()
plt.show()



from sklearn.tree import export_graphviz
import graphviz
import os

# Choose a tree index (e.g., the 5th tree from the 200 estimators)
estimator_index = 4
tree_model = best_model.estimators_[estimator_index]

# Feature names (handle numpy arrays vs. DataFrame)
feature_names = X.columns if hasattr(X, 'columns') else [f'X{i}' for i in range(X.shape[1])]

# Export to DOT format
dot_data = export_graphviz(
    tree_model,
    out_file=None,
    feature_names=feature_names,
    filled=True,
    rounded=True,
    special_characters=True,
    max_depth=3  # limit depth for readability
)

# Render to PNG
graph = graphviz.Source(dot_data)
graph.format = 'png'
graph.render("rfr_tree_example", cleanup=True)

print("Decision tree exported as 'rfr_tree_example.png'")
```
